```python
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import spacy
```

#### Read data

```python
employee_reviews_df = pd.read_csv('employee_reviews.csv')
employee_reviews_df.head()
```

#### Tokenize

```python
def tokenization(word):
    if word is None or (isinstance(word, float)):
        return []
    return word_tokenize(word)
```

```python
employee_reviews_df['review_text_token'] = employee_reviews_df['review_text'].apply(tokenization)
```

#### POS Tagging

```python
nlp = spacy.load('en_core_web_sm')

# preprocessing
token_review = sum(employee_reviews_df['review_text_token'], [])

# POS tagging
pos_doc = nlp(' '.join(token_review))
```

```python
employee_reviews_pos = []

for token in pos_doc:
    employee_reviews_pos.append([token.text, token.pos_])

employee_reviews_pos_df = pd.DataFrame(employee_reviews_pos)
```

```python
employee_reviews_pos_df
```

### NER

```python
employee_reviews_df["ner_doc"] = employee_reviews_df["review_text"].apply(lambda x: nlp(str(x)))
```

```python
ner_rows = []
for idx, doc in employee_reviews_df['ner_doc'].items():
    for ent in doc.ents:
        print(idx, ent.text, ent.label_)
        ner_rows.append([idx, ent.text, ent.label_])

employee_reviews_ner_df = pd.DataFrame(ner_rows)
```
